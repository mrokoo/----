# pandas生成数据
利用DataFrame 和 Series

```python
import pandas as pd ## 导入pandas
import numpy as np ## 导入numpy
```

## 创建数据方法(创建DataFrame)
1. 字典传入，同时可以在frame内部使用series
2. 只要键值对的值能够是一个值或者一个数组偶可以

## 创建series
使用： series() 构造函数。  

	s = pd.Series(data, index=index)

	data可以是Python对象、NumPy的ndarray、一个标量（定值，如8）。index是轴上的一个列表，必须与data的长度相同，如果没有指定，则自动从0开始，表示为[0, …, len(data)–1]。

## 创建DataFrame
  
	基本格式：
	df = pd.DataFrame(data=None, index=None, columns=None, dtype=int32)
	data 是结构化的数据：
	1. 字典
	2. 字典列表，其中每一个字典就是一条记录
	3. Series组成的字典


小结： 
在实际业务中一般不需要我们来生成数据，而是有已经采集好的数据集，直接加载到DataFrame即可。但了解生成数据后可以更有效帮助学习。

```python
df = pd.DataFrame({
	'姓名': pd.Series([23,45,78])
})
s = pd.Series([24, 54, 87, 21, 69], name='gdp')
```
