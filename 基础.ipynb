# 设置索引，即对主要关键字建立索引

```python
df.set_index('name', inplace=True) # 建立索引后表的变化
```

```python
df.head() # 已经将name作为行名。
```

# 数据选取

```python
# 查看指定列
df['Q1'] ## 或者 df.Q1 只要名字符合命名规则就可以
```

```python
# 选择多列
df[['team', 'Q1']] # 只看两列，传入列表
# df.loc[:, ['team', 'Q1']] 与上面效果相同。
```

```python
# 选择行
index = df.index == 'Liver' ## 返回一个索引真值表
df[df.index == 'Liver'] # 通过这个真值表来进行
```

```python
# 自然索引索引选择，类似列表切片
df[0:3] # 取前三行
df[0:10:2] # 0-9 步长：2
df.iloc[:10, :] # 取前10个
```

```python
# 指定行和列
df.loc['Ben', 'Q1':'Q4'] # 只看Ben的四个季度成绩
df.loc['Eorge':'Alexander', 'team':'Q4'] # 指定行列区间
```

```python
# 条件查询
## 单一条件
df[df.Q1 > 90] # Q1列大于90的
df[df.team == 'C'] # team列为C的
df[df.index == 'Oscar'] #索引指定值
## 组合条件
df[(df['Q1'] > 90) & (df['team'] == 'C')] # and关系
df[df['team'] == 'C'].loc[df.Q1 > 90] # 多重筛选
# df[df['team'] == 'C'][df.Q1 > 90] # 多重筛选
```

# 排序

```python
df.sort_values(by='Q1') # 按Q1列数据升序排列，默认按升序排列
df.sort_values('Q1', ascending=False) # 降序 
df.sort_values(['team', 'Q1'], ascending=[True, False]) # team升序，Q1降序
```

# 分组聚类

```python
df.groupby('team').sum() # 按团队分组对应列相加
df.groupby('team').mean() # 按团队分组对应列求平均值
## 不同列不同的计算方法
df.groupby('team').agg({ # 都是对组内的操作
	'Q1': sum, # 总和
	'Q2': 'count',
	'Q3': 'mean',
	'Q4': max
})
```

# 数据转换

```python
df.groupby('team').sum().T
```

# 增加列

```python
df['one'] = 1 # 增加一个固定值的列
df['total'] = df.Q1 + df.Q2 + df.Q3 + df.Q4 # 增加总成绩列
df['total'] = df.loc[:,'Q1':'Q4'].apply(lambda x: sum(x), axis=1)
df['total'] = df.sum(axis=1)
df['avg'] = df.total/4
df
```
