# Numpy基础
Numpy是科学计算包,对数据进行向量化描述，没有任何显式的循环操作，所以执行速度更快，代码更加简洁优雅，出错率更低.
## Numpy的基本对象
+ ndarray: 存储数据的多维数组.
+ ufunc: 对数组进行处理的函数.

ndarray对数据向量化，利用**隐式逐元素方式的广播机制**进行算术运算、逻辑运算、位运算、函数调用等，可以快速、批量地处理数据。
### 创建数据
np.arange() 初、末值 + **步长** 初始化一个线性向量，不包含末值。  
np.linspace() 初、末值 + **元素个数** 初始化一个线性向量，包含末值  
np.random.randn(x,y) 指定形状x × y float64 为数组, 它的分布服从标准高斯分布。
np.random.randint(low, high, sizes=(x,y)) 指定形状x × y int32 为数组, 取值为low -- high之间的整数不包含high
np.zeros(shape,dtype) 生成一个shape(x,y)的零矩阵，多用于矩阵的初始化。
np.ones() 同上，生成一个单位矩阵
np.empty() 同上，不对其初始化
后三者的like，主要为了生成同样形状的矩阵。
np.zero_like() 生成形状相同的零矩阵。

### 矩阵信息
n.shape() 返回矩阵形状
n.shape = (x, y) 改变矩阵形状
更常用的是：
a = n.reshape((x,y)) 生成新数组，防止对原数组的污染。
n.dtype 矩阵数据类型
n.ndim 矩阵维度
n.size 元素数

### 矩阵计算
大多数计算都已经向量化。

```python
np.array([32, 20, 19, 45])[:3]
a = np.array([10, 20, 30, 40])
b = np.linspace(1, 4, 4) 
a + b
a - 1
a / 2
4 * np.log10(a)
a.max()
a.min()
a.sum() # 默认为行和。
```
